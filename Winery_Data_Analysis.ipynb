```python
#caroline parsons
import pandas as p
#open the dataset; caroline parsons
wine = p.read_csv("winemag-data_first150k.csv")
#cleaning up the data by removing the "Unnamed" column from the dataframe; lana Dove
wine.drop('Unnamed: 0', axis=1, inplace=True)
#cleaning up the data to remove columns we do not need for analysis: removing designation; lana Dove
wine.drop('designation', axis=1, inplace = True)
#removing region_ 1 column from dataset; Lana Dove
wine.drop('region_1', axis=1, inplace = True)
#removing region_2 column from dataset; Lana Dove
wine.drop('region_2', axis=1, inplace = True)
#looking at the variables of our wine dataset; Caroline Parsons
wine.head()
```

```python
#drops all the NA from the data set ; Caroline Parsons
winena = wine.dropna()
#looking at the summary statistics for the dataset after dropping NA; Caroline Parsons
winena.describe()
```

```python
# Use the .value_counts() method to get the count of each unique value in the variety column; Caroline Parsons
wine["variety"].value_counts()
#Lana Dove: We have determined to analyize the following: Chardonnay, Pinot Noir, Cabernet Sauvignon, Sauvignon Blanc, Riesling, Merlot, Malbec, Pinot Grigio, and Zinfandel
```

```python
#Group country and show the mean points and price for their wine; Caroline Parsons
wine_byCountry = winena.groupby("country", as_index=False).mean()
wine_byCountry
```
